```python
# Reading an image file with Pillow
from PIL import Image

img = Image.open('image1.jpeg')

img
```

```python
# Let's remove the lower written part from the image
box = (0, 0, 1160, 782)
region = img.crop(box)

region.save('image_1.jpeg')
```

```python
# Splitting and merging bands
img = Image.open('image_1.jpeg')
r,g,b = img.split()

im_new = Image.merge('RGB', (b,g,r))
```

```python
im_new
```

```python
# Transposing the image

img.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
# filtering the image
from PIL import ImageFilter

img.filter(ImageFilter.DETAIL)
```

```python
# Color of the Image
from PIL import ImageColor

img_w = ImageColor.getcolor('white', 'L')
print(img_w)

img_b = ImageColor.getcolor('blue', 'L')
print(img_b)

img_r = ImageColor.getcolor('red', 'L')
print(img_r)
```

```python
# Enhancing the image
from PIL import ImageEnhance

enh = ImageEnhance.Contrast(img)

enh.enhance(1.1)
```

```python
# Adding border to the image
from PIL import ImageOps

border = 20,30,20,50

bmig = ImageOps.expand(im_new, border=border, fill=0)

bmig
```

```python
bmig.size
```

```python
# writing on image
from PIL import ImageDraw, ImageFont

draw = ImageDraw.Draw(bmig)

font = ImageFont.truetype('CherryCreamSoda-Regular.ttf', size=14)

# position of the image
x,y = 1150,860

# message
message = '- Ashrulochan Sahoo'

# color 
color = 'rgb(255,255,255)'

# draw the message on the image
draw.text((1000,830), message, font=font, fill=color)

bmig.show()
```

```python
# Saving the image
bmig.save('modified.png')
```
